```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

        


class BinaryTree:
    def __init__(self):
        self.root = None
    def insert(self, val):
        node = Node(val)
        if not self.root:
            self.root = node
            return
        
        else:
            queue = [self.root]
            while queue:
                temp = queue.pop(0)
                if temp.left is None:
                    temp.left = node
                    return
                queue.append(temp.left)
                
                if temp.right is None:
                    temp.right = node
                    return
                queue.append(temp.right)
                
        return self.root
    
    def search(self, val):
        flag = False
        if self.root:
            queue = [self.root]
            while queue:
                node = queue.pop(0)
                if node.data == val:
                    print(f"The search item : {val} is present in binary tree")
                    return
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                    
        print(f"The search item : {val} is not present in binary tree")
        return
    
    def bfs(self):
        if self.root:
            print("The BFS Traversal of Tree : ", end = "")
            queue = [self.root]
            while queue:
                node = queue.pop(0)
                print(node.data, end = " ")
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            print()
        
    
    def inorder(self, node):
        if node is None:
            return
        self.inorder(node.left)
        print(node.data, end = " ")
        self.inorder(node.right)
                
    def display_inorder(self):
        print("The Inorder Traversal of Tree : ", end = "")
        self.inorder(self.root)
        print()
        
    def preorder(self, node):
        if node is None:
            return
        print(node.data, end = " ")
        self.preorder(node.left)
        self.preorder(node.right)
        
                
    def display_preorder(self):
        print("The Preorder Traversal of Tree : ", end = "")
        self.preorder(self.root)
        print()
                
    def postorder(self, node):
        if node is None:
            return
        self.postorder(node.left)
        self.postorder(node.right)
        print(node.data, end = " ")
                
    def display_postorder(self):
        print("The Postorder Traversal of Tree : ", end = "")
        self.postorder(self.root)
        print()
        
                
    def delete(self, val):
        node = self.root
        target = None
        if node:
            queue = [node]
            while queue:
                temp = queue.pop(0)
                if temp.data == val :
                    target = temp
                    break
                if temp.left:
                    queue.append(temp.left)
                if temp.right:
                    queue.append(temp.right)

        if not target:
            print(f"Node => ({val}) to be deleted is not present in list ....")
            return
        else:
            print("Data to be deleted : ", target.data)
        
        last_node = None
        last_parent = None
        
        queue = [(self.root, None)]
        
        while queue:
            curr, parent = queue.pop(0)
            last_node = curr
            last_parent = parent
            if curr.left:
                queue.append((curr.left, curr))
            if curr.right:
                queue.append((curr.right, curr))
        
        target.data = last_node.data
        
        if last_parent:
            if last_parent.left == last_node:
                last_parent.left = None
            else:
                last_parent.right = None
            
        print(f"The item : {val} is deleted from binary tree")
        return
                
                    
            
        
```

```python
tree = BinaryTree()
tree.insert(10)
tree.insert(8)
tree.insert(20)
tree.insert(16)
tree.insert(18)
tree.insert(22)
tree.insert(24)
tree.bfs()
tree.display_inorder()
tree.display_preorder()
tree.display_postorder()
tree.search(28)
tree.search(24)
tree.delete(24)
tree.display_inorder()
tree.display_preorder()
tree.display_postorder()
tree.search(24)
tree.delete(28)
```
